### 创建图 -- networkX 的内置图

#### 经典图结构


```python
# 全连接无向图
G = nx.complete_graph(7)
nx.draw(G) # 画图
G.size() # 全图连接数
```

```python
# 全连接有向图 
G = nx.complete_graph(7, nx.DiGraph())
nx.draw(G)
G.is_directed()
```

```python
# 环状图
G = nx.cycle_graph(5)
nx.draw(G)
```

```python
# 梯状图
G = nx.ladder_graph(5)
nx.draw(G)
```

```python
# 线性串珠图
G = nx.path_graph(15)
nx.draw(G)
```

```python
# 星状图
G = nx.star_graph(7)
nx.draw(G)
```

```python
# 轮辐图
G = nx.wheel_graph(8)
nx.draw(G)
```

```python
# 二项树
G = nx.binomial_tree(5)
nx.draw(G)
```

```python
# 二维矩形网格图
G = nx.grid_2d_graph(3,5)
nx.draw(G)
```

```python
# 多维矩形网格图
G = nx.grid_graph(dim=(2,3,4))
nx.draw(G)
```

```python
# 二维三角形网格图
G = nx.triangular_lattice_graph(2,5)
nx.draw(G)
```

```python
# n维超立方体图
G = nx.hypercube_graph(4)
nx.draw(G)
```

```python
# networkX 内置图
G = nx.diamond_graph()
nx.draw(G)
```

```python
G = nx.bull_graph()
nx.draw(G)
```

```python
# 无标度有向图
G = nx.scale_free_graph(5)
nx.draw(G)
```

```python
# 空手道俱乐部数据集
G = nx.karate_club_graph()
nx.draw(G, with_labels=True)
```

```python
G.nodes[5]["club"]
G.nodes[9]["club"]
```

```python
# 雨果《悲惨世界》人物关系
import matplotlib.pyplot as plt 
G = nx.les_miserables_graph()
plt.figure(figsize=(12,16))
pos = nx.spring_layout(G, seed=12)
nx.draw(G, pos, with_labels=True)
```

```python
# Florentine families graph 
G = nx.florentine_families_graph()
nx.draw(G, with_labels=True)
```

```python
# 社群聚类图
G = nx.caveman_graph(4,3)
nx.draw(G, with_labels=True)
```

```python
# tree 
tree = nx.random_tree(n=10, seed=0)
print(nx.forest_str(tree,sources=[0]))
```
